## Pegando os dados da coluna ```Valor``` é usando o comando ```split()``` para separar em ```array``` e posteriormente tratar os dados.

####O comando ```dados.Valor.str.split()``` gerou uma series com uma separação de todas as linhas agora a partir do comando ```split()```.

Se rodar o comando ```dados.Valor.str.split()``` sem o ```str``` gera erro. É necessário informar que estamos aplicando o método de string.

````python
dados.Valor.str.split()
````

####```dados.Valor.str.split(expand = True)``` esse comando passando o parâmetro ```expand = True``` separa os dados em colunas. Dessa maneira, torna a visualização melhor dos dados e nos permite analisar e tratar os dados da melhor maneira.

````python
dados.Valor.str.split(expand = True)
````

####O comando ```dados.Valor.str.split(expand = True)[0].unique()``` passando entre ```[]``` o zero (signfica a coluna zero) é chamando a função ```unique()```  nos retorna os valores que possui nessa coluna.

````python
dados.Valor.str.split(expand = True)[0].unique()
````

````python
dados.Valor.str.split(expand = True)[2].unique()
````

````python
dados.Valor.str.split(expand = True)
````

####Passando as colunas que foram geradas através do ```split(expand = True)``` para o DataFrame ```dados```. 

````python
dados[["Moeda", "Valor_anuncio", "Tipo_anuncio"]] = dados.Valor.str.split(expand = True)
dados.head()
````

####Visualizando os tipos de dados de cada coluna.

````python
dados.info()
````

#### Remover os dados da coluna ```Tipo_anuncio``` que **não é vazio**.
OBS: Esses dados são os imóveis à venda!

````python
dados.Tipo_anuncio.isnull()
````

####Visualizando a remoção.

OBS: O DataFrame ```dados``` ainda contém os dados não nulos.

````python
dados[dados.Tipo_anuncio.isnull()]
````

#### Visualizando se o comando acima funcionou.

#### Salvando em uma variável ```dados_vendas```. 
Essa variável contém as informações de todos os imóveis que estãos à venda. 

A váriavel ```dados``` continua com as dez mil linhas.

````python
dados_vendas = dados[dados.Tipo_anuncio.isnull()]
dados_vendas
````

````python
dados
````

#### Alterando o tipo de dado do campo ```Valor_anuncio```.

1. Remove os pontos que tem nos números, pois o pandas entende que o ponto é casa decimal, e como tem dois pontos nesses números ele da conflito.
2. Altera o tipo do campo para ```float```.
3. Cria uma coluna `dados_vendas["Valor_anuncio_float"]`

````python
dados_vendas["Valor_anuncio"].str.replace(".", "").astype(float)
````

````python
dados_vendas["Valor_anuncio_float"] = dados_vendas["Valor_anuncio"].str.replace(".", "").astype(float)
````

````python
dados_vendas.info()
````

#### Criando um histograma

````python
dados_vendas.Valor_anuncio_float.plot.hist(bins = 200)
````

#### Visualizando informações com o describe()

````python
dados_vendas.describe()
````

````python
pd.set_option("display.precision", 2) #Coloca duas casas decimais
pd.set_option("display.float_format", lambda x: '%.2f' % x)
dados_vendas.describe()
````

````python
dados_vendas.describe(include = 'all')
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set() #estilizar
plt.figure(figsize=(10,8)) #configura o tamanho do gráfico
ax = sns.histplot(data = dados_vendas, x = 'Valor_anuncio_float', kde = True) #kde = true faz uma curva no histograma
ax.set_title("Histograma Valor imóvel")
plt.xlim(-50, 10000000)
plt.show()
````
